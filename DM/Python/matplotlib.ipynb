### q1

```python
import matplotlib.pyplot as plt

x = range(1,50);
y = [t * 3 for t in x]
print("X: ", *x)
print("Y: ", y)
plt.plot(x, y)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Draw a line")
```

### q2

```python
import numpy as np

x = range(1, 4)
y = []
for t in x:
    if t < 2:
        y.append(t*2)
    else:
        y.append(-3 * t + 10)
print("X: ", *x)
print("Y: ", y)
plt.plot(x, y)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Graph")
```

### q5

```python
x1 = range(10, 31)
y1 = []
for t in x1:
    if t <= 20:
        y1.append(-3 * t + 70)
    else:
        y1.append(2 * t - 30)
plt.plot(x1, y1, color = 'red', linewidth = 5, label ="line width = 5")
x2 = range(10, 31)
y2 = []
for t in x2:
    if t <= 20:
        y2.append(2 * t)
    else:
        y2.append(-3 * t + 100)
plt.plot(x2, y2, color = 'blue', linewidth =3, label ="line width = 3")
plt.legend()
```
